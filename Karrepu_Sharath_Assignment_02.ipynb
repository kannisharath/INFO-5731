```python
import pandas as pd
import warnings
from semanticscholar import SemanticScholar
# Ignoring warnings
warnings.filterwarnings("ignore")
pd.set_option('display.width', None)
# Initialization
sem_sch = SemanticScholar()
# Search articles using "any string"
res = sem_sch.search_paper('data science', fields=['title', 'abstract'])
# Create dataframe
df_paper = pd.DataFrame(columns=['paperId', 'title', 'abstract'])
# Iterate through the search results and add each paper's information to the DataFrame
count = 0
for i in res:
    count += 1
    paper_info = {
        'paperId': i.paperId,
        'title': i.title,
        'abstract': i.abstract
    }
    df_paper = df_paper.append(paper_info, ignore_index=True)

    if count == 10000:
        break
# Save the DataFrame to a CSV file
df_paper.to_csv('papers_data_science.csv', index=False)
# Print the DataFrame
print(df_paper)
```

# Question 2 (30 points)

Write a python program to **clean the text data** you collected in the previous question and save the clean data in a new column in the csv file. The data cleaning steps include: [Code and output is required for each part]

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the stopwords list.

(4) Lowercase all texts

(5) Stemming.

(6) Lemmatization.

```python
# Write code for each of the sub parts with proper comments.
!pip install pandas nltk

```

```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import nltk

# Ensure you have downloaded the necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Load the data
df = pd.read_csv('papers_data_science.csv')

# Remove noise, such as special characters and punctuations.
df['clean_abstract'] = df['abstract'].apply(lambda x: re.sub(r'[^\w\s]', '', str(x)))

#  Remove numbers
df['clean_abstract'] = df['clean_abstract'].apply(lambda x: re.sub(r'\d+', '', x))

# Get English stop words
stop_words = set(stopwords.words('english'))

# Remove stop words from the 'clean_abstract' column
df['clean_abstract'] = df['clean_abstract'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stop_words]))

# Convert  to lowercase (Stemming)
df['clean_abstract'] = df['clean_abstract'].apply(lambda x: x.lower())

# Initialize Porter Stemmer
stemmer = PorterStemmer()

# Apply stemming to 'clean_abstract' column
df['clean_abstract'] = df['clean_abstract'].apply(lambda x: ' '.join([stemmer.stem(word) for word in x.split()]))

# Print the head of the DataFrame
print(df.head())

# Save the remaining data to a CSV file
df.iloc[5:].to_csv('cleaned_text_data.csv', index=False)
#In the below output we can see the differnce between abstract and cleaned abstract
```

# Parts of Speech (POS) Tagging:

```python
import pandas as pd
import nltk

# Ensure you have downloaded the necessary NLTK data
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Function for Parts of Speech (POS) Tagging
def pos_tagging(text):
    tokens = nltk.word_tokenize(text)
    tagged = nltk.pos_tag(tokens)

    # Initialize counters for each POS tag
    noun_count = 0
    verb_count = 0
    adj_count = 0
    adv_count = 0

    for word, tag in tagged:
        if tag.startswith('N'):  # Noun
            noun_count += 1
        elif tag.startswith('V'):  # Verb
            verb_count += 1
        elif tag.startswith('J'):  # Adjective
            adj_count += 1
        elif tag.startswith('R'):  # Adverb
            adv_count += 1

    return noun_count, verb_count, adj_count, adv_count

# Load the clean text from the CSV file and handle NaN values
df_clean = pd.read_csv('cleaned_text_data.csv')
df_clean['clean_abstract'] = df_clean['clean_abstract'].fillna('')  # Fill NaN values with an empty string
clean_text = ' '.join(df_clean['clean_abstract'])

# Perform POS tagging on the clean text
noun, verb, adj, adv = pos_tagging(clean_text)
print("Noun count:", noun)
print("Verb count:", verb)
print("Adjective count:", adj)
print("Adverb count:", adv)
```

```python
# 2)
import pandas as pd
import spacy

# Load English tokenizer, tagger, parser, NER, and word vectors for SpaCy
nlp = spacy.load("en_core_web_sm")

# Load the clean text from the CSV file and handle NaN values
df_clean = pd.read_csv('cleaned_text_data.csv')
df_clean['clean_abstract'] = df_clean['clean_abstract'].fillna('')  # Fill NaN values with an empty string
clean_text = ' '.join(df_clean['clean_abstract'])

# Function for Constituency Parsing and Dependency Parsing
def parsing(text):
    doc = nlp(text)

    # Print out constituency parsing tree
    print("Constituency parsing tree:")
    for token in doc:
        print(token.text, token.dep_, token.head.text, token.head.pos_,
                [child for child in token.children])

    # Print out dependency parsing tree
    print("\nDependency parsing tree:")
    for chunk in doc.noun_chunks:
        print(chunk.text, chunk.root.text, chunk.root.dep_,
                chunk.root.head.text)

# Perform parsing on the clean text
parsing(clean_text)
```

```python
# 3)
import pandas as pd
import spacy

# Load English tokenizer, tagger, parser, NER, and word vectors for SpaCy
nlp = spacy.load("en_core_web_sm")

# Function for Named Entity Recognition (NER)
def named_entity_recognition(text):
    entities = {}
    doc = nlp(text)

    for ent in doc.ents:
        if ent.label_ not in entities:
            entities[ent.label_] = []
        entities[ent.label_].append(ent.text)

    return entities

# Load the clean text from the CSV file and handle NaN values
df_clean = pd.read_csv('cleaned_text_data.csv')
df_clean['clean_abstract'] = df_clean['clean_abstract'].fillna('')  # Fill NaN values with an empty string
clean_text = ' '.join(df_clean['clean_abstract'])

# Perform Named Entity Recognition on the clean text
entities = named_entity_recognition(clean_text)
for label, values in entities.items():
    print(label, ":", values)
```
